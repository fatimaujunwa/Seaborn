## Data Visualization with seaborn ðŸ“Š

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("exams.csv")
data
```

## Lineplot

```python
# they basically show frequency of each data point
plt.figure(figsize=(14,6))
plt.title("ploting the whole data on line plot")
sns.set_style("darkgrid")
# "darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"
sns.lineplot(data=data)
#hi this is a
```

```python
plt.figure(figsize=(14,6))
plt.title("ploting data points for reading",fontsize=18)
list(data.columns)
sns.lineplot(data=data["reading"])
plt.ylabel("Reading",fontsize=18)
plt.xlabel('Count',fontsize=18)
```

### Barplot

```python
plt.figure(figsize=(14,6))
plt.title("Bar plot")
sns.barplot(data=data)

#sns wont plot categorical variables on any type of plots
#automatically plots the discrete variables and the index
```

```python
# ploting a subset of the data writing

plt.figure(figsize=(10,10))
plt.title("Bar plot")
sns.barplot(y=data["writing"],x=data.index)
```

## heatmaps

```python
plt.figure(figsize=(10,10))
plt.title("HeatMaps")
reading=data[["reading"]].head(10)
data_head=data.head(10)
sns.heatmap(data=reading,annot=True)
#it has to not be flat to plot on heat map
plt.xlabel("count")
```

### Scatter plots

```python
plt.figure(figsize=(5,5))
plt.title("Scatter plot of writing and reading")
sns.scatterplot(x=data["writing"],y=data["reading"],hue=data['test_prep_course'])
# The scatterplot above suggests that body mass index (BMI) and insurance charges are positively correlated,
# this plot basically says those with no completed courses still have a high positive corr in their reading and writing as much as those who completed
#means their reading scores increase with increasing writing whether they completed the prep scores or not
```

### regression plot

```python
sns.regplot(x=data["writing"],y=data['reading'])
```

```python
sns.lmplot(x="writing",y="reading",data=data,hue='test_prep_course')
```

```python
sns.swarmplot(x=data['test_prep_course'],y=data['reading'])
#this shows that for reading those who completed the prep course got lower scores in reading that those who completed in prep course
```

```python
sns.histplot(data['writing'],bins=8)
```

```python
# kernel density estimate is a smooth plot
sns.kdeplot(data=data['writing'],shade=True)
```

```python
sns.histplot(data=data,x='writing',hue='test_prep_course')
#this shows the frequency of writing data set for those that completed and did not complete the test
#the height of histogram shows how many valiues fall into that bin range
```
